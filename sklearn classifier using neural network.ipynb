```python
df=df.drop(['texture_mean','perimeter_mean','area_mean','radius_worst'],axis=1)
```

```python
categorical=[]
for i in df:
    if df[i].dtypes =='object':
        categorical.append(i)
print(categorical)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for i in categorical:
    df[i]=le.fit_transform(df[i])
```

```python
df.head()
```

```python
target=df.iloc[:,0]
```

```python
numeric=[]
for i in df:
    if df[i].dtypes!='object':
        numeric.append(i)
```

```python
predictor=df[numeric]
```

```python
predictor.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(predictor,target,test_size=0.2,random_state=42)
```

```python
sc.fit(x_train)
```

```python
x_train=sc.transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
ml=MLPClassifier()
```

```python
ml.fit(x_train,y_train)
```

```python
prediction=ml.predict(x_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test,prediction))
```

```python
print(metrics.accuracy_score(y_test,prediction))
```

```python
print(metrics.confusion_matrix(y_test,prediction))
```

```python
ml=MLPClassifier(hidden_layer_sizes=(10,10,10),activation='relu',solver='lbfgs',alpha=0.001)
```

```python
ml.fit(x_train,y_train)
```

```python
prediction=ml.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print(metrics.classification_report(y_test,prediction))
print(metrics.confusion_matrix(y_test,prediction),end=' ')
print(metrics.accuracy_score(y_test,prediction), ' accuracy')
```

```python
ml=MLPClassifier(hidden_layer_sizes=(10,15,20),activation='relu',solver='sgd',learning_rate='adaptive',alpha=0.001,max_iter=300,random_state=42)
```

```python
ml.fit(x_train,y_train)
```

```python
prediction=ml.predict(x_test)
```

```python
print(metrics.classification_report(y_test,prediction))
print(metrics.confusion_matrix(y_test,prediction),end=' ')
print(metrics.accuracy_score(y_test,prediction), ' accuracy')
```
